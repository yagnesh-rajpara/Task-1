```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')

fig, axes = plt.subplots(2, 3, figsize=(14, 10))
fig.suptitle('TITANIC DATASET - PRODUCTION EDA DASHBOARD', 
             fontsize=16, fontweight='bold', y=0.95)

# 1. Target Distribution
df['Survived'].value_counts(normalize=True).plot(kind='bar', ax=axes[0,0], 
                                                 color=['#d62728', '#2ca02c'])
axes[0,0].set_title('Target Imbalance\n(38.4% Survival Rate)', fontweight='bold')
axes[0,0].set_ylabel('Proportion')
axes[0,0].tick_params(axis='x', rotation=0)

# 2. Age vs Survival 
age_data = df.dropna(subset=['Age'])
axes[0,1].boxplot([age_data[age_data['Survived']==0]['Age'], 
                   age_data[age_data['Survived']==1]['Age']],
                 labels=['Not Survived', 'Survived'])
axes[0,1].set_title('Age Distribution by Survival', fontweight='bold')
axes[0,1].set_ylabel('Age (years)')

# 3. Pclass Survival
pclass_surv = df.groupby('Pclass')['Survived'].mean()
pclass_surv.plot(kind='bar', ax=axes[0,2], color=['#ff7f0e', '#2ca02c', '#d62728'])
axes[0,2].set_title('Survival Rate by Passenger Class', fontweight='bold')
axes[0,2].set_ylabel('Survival Rate')

# 4. Gender Survival
sex_surv = df.groupby('Sex')['Survived'].mean()
sex_surv.plot(kind='bar', ax=axes[1,0], color=['#1f77b4', '#ff7f0e'])
axes[1,0].set_title('Survival Rate by Gender', fontweight='bold')
axes[1,0].set_ylabel('Survival Rate')

# 5. Fare Distribution
df['Fare'].hist(bins=30, ax=axes[1,1], alpha=0.7, color='#2ca02c', edgecolor='black')
axes[1,1].set_title('Fare Distribution\n(Highly Right-Skewed)', fontweight='bold')
axes[1,1].set_xlabel('Fare ($)')
axes[1,1].set_ylabel('Frequency')

# 6. Missing Values
missing_pct = df.isnull().sum().sort_values(ascending=False) / len(df) * 100
missing_pct[missing_pct > 0].plot(kind='barh', ax=axes[1,2], color='salmon')
axes[1,2].set_title('Missing Values %', fontweight='bold')
axes[1,2].set_xlabel('Missing Percentage')

plt.tight_layout()
plt.savefig('titanic_eda_dashboard.png', dpi=300, bbox_inches='tight', 
            facecolor='white')
plt.show()
```
